```python
import numpy as np
import pandas as pd
```

```python
def binary_pixel_convert(data_frame:pd.DataFrame, margin=0) -> pd.DataFrame:
    sub_df=data_frame.copy()
    for col in data_frame.columns:
      sub_df.loc[sub_df[col]<=margin, col]=0
      sub_df.loc[sub_df[col]>margin, col]=1
    return sub_df
```

```python
MARGIN=10
train_file_path = '/content/drive/MyDrive/cyberlabs_google_colab/Colab Notebooks/emnist-letters-train.csv'
train_df = pd.read_csv(train_file_path)
x_train = train_df.iloc[:, 1:785]
x_train = binary_pixel_convert(x_train, margin=MARGIN)
x_train = np.array(x_train)
y_input = np.array(train_df[['23']])
```

```python
def Euclidean_distance(x1: np.array, x2: np.array) -> float:
    value = np.sum((x1-x2)**2, axis=1)
    return np.sqrt(value)

def Hamming_distance(x1: np.array, x2: np.array) -> int:
    x_diff = x1-x2
    value = np.count_nonzero(x_diff, axis=1)
    return value

def Manhattan_distance(x1: np.array, x2: np.array) ->float:
    value = np.sum(abs(x1-x2), axis=1)
    return value
```
