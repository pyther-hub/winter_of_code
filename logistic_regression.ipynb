```python
import pandas as pd
import numpy as np
file_path = '/content/drive/MyDrive/cyberlabs_google_colab/Colab Notebooks/emnist-letters-train.csv'
df=pd.read_csv(file_path)
```

```python
def standardize(feature_value: np.array):
    value = (feature_value-np.mean(feature_value))/np.std(feature_value)
    return value

def sigmoid(array: np.array):
    value = 1/(1+(np.exp(-array)))
    return value

def cost(y_true:np.array, y_pred:np.array):
    # print(y_pred, max(y_pred))
    # print(y_true, max(y_true))
    p1 =np.sum(y_true*np.log(y_pred))
    p2 = np.sum((1-y_true)*np.log(1-y_pred))
    cost_value=-(p1+p2)
    return cost_value
```

```python
def get_para(x_input, y_train):
  learning_rate=0.5
  max_iter=1000
  numb_rows, numb_featres = x_input.shape
  parameters = np.zeros((numb_featres, 1))
  iter_pos = 0
  bias=1
  while iter_pos < max_iter:
      predicted_value=np.dot(x_input, parameters)+bias
      activation_value = sigmoid(predicted_value)
      error = activation_value-y_train 
      matrix_prod_grad_desc = np.dot(x_input.T,error)
      parameters -= matrix_prod_grad_desc*(learning_rate/numb_rows)
      bias -= np.sum(error)*(learning_rate/numb_rows)
      iter_pos += 1
  return parameters, bias
```

```python
def accuracy(y_true:np.array, y_predicted:np.array):
    n=y_predicted.shape[0]
    correct=0
    wrong=0
    for pos in range(n):
        if y_predicted[pos]==y_true[pos]:
            correct+=1
        else:
            wrong+=1
    return (correct/(wrong+correct))


```

```python
all_parameters = {}
all_data_frames=[]
for output in range(1,27):
    print(output)
    sub_df=df.copy()
    sub_df.loc[df['23']!=output, '23']=0
    sub_df.loc[df['23'] == output, '23'] = 1
    x_input=np.array(sub_df.iloc[:,1:785])/255
    y_true=np.array(sub_df[['23']])
    parameter_value, bias=get_para(x_input, y_true)
    all_parameters[output]=(parameter_value, bias)
    y_pred=sigmoid(np.dot(x_input, parameter_value)+bias)
    print('--'*10)
```

```python
import pickle 
f=open('/content/drive/MyDrive/cyberlabs_google_colab/Colab Notebooks/all_parameters.dat', 'wb')
pickle.dump(all_parameters, f)
f.close()
```

```python
row_numb=0
x_input=np.array(df.iloc[:,1:785])/255
y_train=np.array(df[['23']])
correct_predictions=0
wrong_predictions=0
for row in x_input:
  activations=[]
  row=np.atleast_2d(row)
  for pos in range(26):
    para, b=all_parameters[pos+1]
    pred_value=sigmoid(np.dot(row,para)+b)
    activations.append(pred_value)
  letter=activations.index(max(activations))+1
  true_letter=y_train[row_numb]
  if letter==true_letter:
    correct_predictions+=1
  else:
    wrong_predictions+=1
  row_numb+=1
print(correct_predictions/(wrong_predictions+correct_predictions))
```
