```python
import numpy as np
import pandas as pd
```

```python
def sigmoid(array:np.array) -> np.array:
    return 1/(1+np.exp(-array))

def generate_parameters():
    parameters={}
    for pos in range(n_layers-1):
        parameters[f'w{pos+1}'] = np.random.rand(nodes[pos+1], nodes[pos])-0.5
        parameters[f'b{pos+1}'] = np.random.rand(nodes[pos+1], 1)-0.5
    return parameters
```

```python
def forward_prop(x_input, parameter):
    a=x_input
    nodes_values[f'l{1}']=a
    print('FORWARD PASS'+'-'*10)
    for pos in range(1,n_layers):
        w = parameter[f'w{pos}']
        b=parameter[f'b{pos}']
        z=np.dot(w,a)+b
        a=sigmoid(z)
        print(f'z{pos+1}=w{pos}.a{pos}+b{pos}')
        print(f'a{pos+1}=sigmoid(z{pos+1})')
        nodes_values[f'l{pos+1}']=a
    print('forward pass ends ----')
    return a
```

```python
def back_prop(parameters, y_true, y_pred):
    delta = (nodes_values[f'l{n_layers}']-y_true)
    delta_para={}
    print('-'*4+'BACK PROP STARTS'+'-'*6)
    print(f'delta{n_layers}=l{n_layers}-y_true')

    delta_para[f'delW{n_layers-1}']=np.dot(delta, nodes_values[f'l{n_layers-1}'].T)*lr/m
    delta_para[f'delB{n_layers-1}']=(np.sum(delta, axis=1, keepdims=True))*lr/m

    print(f'dW{n_layers-1}=(delta{n_layers}.a{n_layers-1}T)*lr/m')
    print(f'db{n_layers-1}=(sum(delta{n_layers}))*lr/m')

    for pos in range(n_layers-1,1,-1):
        a=nodes_values[f'l{pos}']
        x=nodes_values[f'l{pos-1}']
        delta=np.dot(parameters[f'w{pos}'].T, delta)*((a)*(1-a))

        print(f'delta{pos}=(w{pos}T).delta{pos+1}x(a{pos})*(1-a{pos})')

        delta_para[f'delW{pos-1}']=np.dot(delta, x.T)*lr/m
        delta_para[f'delB{pos-1}']=np.sum(delta, axis=1, keepdims=True)*lr/m

        print(f'dW{pos-1}=delta{pos}.a{pos-1}T')
        print(f'dB{pos-1}=delta{pos}.ones')
    print('BACK PROP ENDSSS', 'X'*9)
    return delta_para
```

```python
def update_parameters(parameters, delta_para):
    for pos in range(n_layers-1,0,-1):
        parameters[f'w{pos}']-=delta_para[f'delW{pos}']
        parameters[f'b{pos}']-=delta_para[f'delB{pos}']
    return parameters

```

```python
def cost(y_hat, y):
    value=y*np.log(y_hat)+(1-y)*np.log(1-y_hat)
    cost_val=np.sum(value)
    return -cost_val/m
```

```python
global nodes_values
nodes=[784,30,20,26]
lr=0.3
nodes_values = {}
n_layers=len(nodes)
```

```python
file_path = '/content/drive/MyDrive/cyberlabs_google_colab/Colab Notebooks/emnist-letters-train.csv'
df=pd.read_csv(file_path)
x_train=df.iloc[:,1:785]
y_train=df[['23']]
m=y_train.shape[0]
x_train = np.array(x_train).T/255
y_train = np.array(y_train)
```

```python
multi_class_y_train=[]
for element in y_train:
    element = int(element)
    value = (0,)*(element-1)+(1,)+(0,)*(26-element)
    multi_class_y_train.append(value)
multi_class_y_train = np.array(multi_class_y_train).T
```

```python
iter_pos=0
max_iter=2
iter_cost=[]
parameters=generate_parameters()
while iter_pos<max_iter:
    y_pred = forward_prop(x_train, parameters)
    cost_value = cost(y_hat=y_pred, y=multi_class_y_train)
    iter_cost.append(cost_value)

    delta_para = back_prop(parameters, multi_class_y_train, y_pred)
    parameters=update_parameters(parameters,delta_para)

    print('--0'*(20))
    print(iter_pos, cost_value)
    iter_pos+=1
    
```
